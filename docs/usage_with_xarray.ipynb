# Extending xarray

The `ssnolib` provides a xarray extension, which allows addressing xarray objects by their standard name:

```python
import ssnolib
from ssnolib import xraccessor
import xarray as xr
```

```python
ds1 = xr.Dataset(
    {"data": xr.DataArray([[1, 2], [3, 4]], dims=("x", "y"))},
    coords={"x": [1, 2], "y": [3, 4]},
)
```

```python
ds1
```

```python
str(ssnolib.StandardName(standardName="x_coordinate", canonicalUnits="m", description="The x coordinate"))
```

```python
ds1.x.attrs["standard_name"] = ssnolib.StandardName(standardName="x_coordinate", canonicalUnits="m", description="The x coordinate")
```

```python
ds1.x.attrs["standard_name"]
```

```python
ds1.from_dict(ds1.to_dict()).x.attrs["standard_name"]
```
